# 1. Delete redundant columns


```python
data.columns  # List all columns in the dataset
```

```python
columns_to_keep = ['name', 'online_order', 'book_table', 'rate', 'dish_liked', 'approx_cost(for two people)']
columns_to_drop = ['url', 'address', 'votes', 'phone', 'location', 'rest_type', 'cuisines', 'reviews_list', 'menu_item', 'listed_in(type)', 'listed_in(city)']
```

# 2A. Drop the columns

```python
columns_to_drop
```

```python
data.drop(columns=columns_to_drop, inplace=True)  # Drop the specified columns from the DataFrame
```

# 2B. Rename the columns

```python
data
```

```python
data.columns  # List all columns in the dataset
```

```python
for i in data.columns:
    print(i.capitalize())  # Capitalize the first letter of each column name
```

```python
new_column_names = []
for i in data.columns:
    new_column_names.append(i.capitalize())  # Capitalize the first letter of each column name
```

# 3. Remove duplicates

```python
data.duplicated().sum()   # Check for duplicates in the DataFrame
```

```python
data.drop_duplicates(inplace=True)  # Remove duplicate rows from the original DataFrame
```

```python
data.duplicated().sum() # Check again for duplicates to ensure they have been removed
```

# 4. Remove the NaN values from the dataset

```python
data.isna().sum()  # Check for NaN values in the DataFrame
```

```python
data.dropna(inplace=True)  # Remove rows with NaN values from the DataFrame
```

```python
data.isna().sum()  # Check again for NaN values to ensure they have been removed
```

```python
data
```

```python
data.reset_index(inplace=True)  # Reset the index of the DataFrame after cleaning
```

```python
data
```

```python
data
```

# 5. Clean individual columns

```python
data['Rate'][14]
```

```python
data['Rate']
```

```python
data['Rate'] = data['Rate'].str.replace(' ', '')  # Replace ' ' with '' in the 'Rate' column
```

```python
data
```

```python

data['Rate'] = data['Rate'].str.replace('/', ' out of ')  # Replace '/' with ' out of ' in the 'Rate' column
```

```python
data
```

# 6. Check for some more transformation

```python
data
```

```python
data['Online_order']
```

```python
data['Online_order'] = data['Online_order'].apply(lambda x: 1 if x == 'Yes' else 0)  # Convert 'Yes'/'No' to 1/0 in the 'Online_order' column
```
